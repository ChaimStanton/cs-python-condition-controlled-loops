## Task 4: The sequence 1,4,9,16,25 is made up of square numbers. Write a program that writes out all the square numbers under 500.

````python
# write your code here

def squareNum(numberToPrintTill):
    counter = 1 
    numberToPrintTill = numberToPrintTill + 1
    while counter < numberToPrintTill:
        print(counter, "squared = ", counter**2)
        counter += 1

# now Call the function 
squareNum(500)
````

## Task 5. Write a program to simulate throwing a die until 6 is thrown. The program outputs the average number of throws it took to get 6

````python
# why do we need to use random library? Because python does not have the random function built in by defult 
from random import randint
totalThrows = 0
dieResult = 0
while dieResult != 6:
    dieResult = randint(1,6)
    totalThrows += 1

print("It took", totalThrows, "throws to get a 6")
````

## Extension task
Scrabble is a game where players get points by spelling words. Words are scored by adding together the point values of each individual letter.Define a function scrabble_score that takes a string word as input and returns the equivalent scrabble score for that word.

````python
letterScore = {"a": 1, "c": 3, "b": 3, "e": 1, "d": 2, "g": 2,
               "f": 4, "i": 1, "h": 4, "k": 5, "j": 8, "m": 3,
               "l": 1, "o": 1, "n": 1, "q": 10, "p": 3, "s": 1,
               "r": 1, "u": 1, "t": 1, "w": 4, "v": 4, "y": 4,
               "x": 8, "z": 10}


# write your code inside the function
def scrabble_score(word):
    score = 0
    counter = 0
    while counter <= len(word) - 1:
        letter = word[counter].lower()
        score = score + letterScore[letter]
        counter += 1
    return score


# Call the function with input parameter "Hello"
print("your scrabble score is", scrabble_score("Hello"))

````
